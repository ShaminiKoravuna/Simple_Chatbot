# Fetching the model and initializing a tokenizer

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/blenderbot-400M-distill"

tokenizer = AutoTokenizer.from_pretrained(model_name) # interacting with the model

model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

# Encoding conversation history and prompt from the user

```python
conversation_history = []

history_string = "\n".join(conversation_history)
```

```python
input_text = "Hello!"
```

# Tokenization of user prompt and chat history

```python
inputs = tokenizer.encode_plus(history_string, input_text, return_tensors="pt")
```

```python
inputs
```

# Generating output from the model

```python
outputs = model.generate(**inputs)
```

```python
outputs
```

# Decoding the output

```python
response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
```

```python
response
```

# Updating the conversation history

```python
conversation_history.append(input_text)
conversation_history.append(response)
```

```python
conversation_history
```

# Repeating the process

```python
history_string = "\n".join(conversation_history)
```

```python
input_text = "See you later!"
```

```python
inputs = tokenizer.encode_plus(history_string, input_text, return_tensors="pt")
```

```python
outputs = model.generate(**inputs)
```

```python
response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
```

```python
response
```

```python
conversation_history.append(input_text)
conversation_history.append(response)
```
